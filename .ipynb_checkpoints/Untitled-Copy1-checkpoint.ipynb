```python
import wfBm
import scipy.stats as stats
import matplotlib.pyplot as plt, mpld3

corr = 0
v1 = []
v2 = []
values1 = wfBm.wfBm()
values2 = wfBm.wfBm()
v1 = values1
v2 = values2
last1 = values1[-1]
last2 = values2[-1]
for i in range(3):
    while abs(corr) < 0.85:
        values1 = wfBm.wfBm()
        values2 = wfBm.wfBm()
        corr, p = stats.pearsonr(values1, values2)
    first1 = values1[0]
    first2 = values2[0]
    shift1 = last1 - first1
    shift2 = last2 - first2
    values1 = [v + shift1 for v in values1]
    values2 = [v + shift2 for v in values2]
    v1.extend(values1)
    v2.extend(values2)
    last1 = values1[-1]
    last2 = values2[-1]
    corr = 0

plt.title('fBm (blue) vs BM (red)')
plt.xlabel('i')
plt.ylabel('x(i)')
plt.scatter(v1,v2)
plt.plot(v1,v2)
mpld3.display()
```
